The Gumbel-Max trick for discrete distributions
===============================================

This note was inspired by: https://hips.seas.harvard.edu/blog/2013/04/06/the-gumbel-max-trick-for-discrete-distributions/

The problem: We want to generate data from the following distribution

$$
  \pi_k = \exp(x_k) / Z
$$

This parameterization is known as the "natural parameterization" of a discrete distribution

Want to do this without applying the softmax transform (i.e. no calls to $exp(x_k)$ or a big summation to compute $Z$) and doing the usual thing to sample from a discrete distribution.

It turns out that the following trick is equivalent to the softmax-discrete
procedure: add Gumbel noise to each and then take the \argmax. That is, add
independent Gumbel noise $z_k$ to each $x_k$ and then $\text{argmax}_k \ x_k + z_k$ is
an unbiased sample from $p$.


```python
import pylab as pl
import numpy as np
import scipy.stats as st
from numpy import log, exp
from numpy.random import gumbel, uniform
from arsenal.timer import Timer
from arsenal.math import exp_normalize, logsumexp
```

Here's an interesting observation:

The Gumbel trick never computes $Z$! This means that we can start working on our samples before we've even seen the entire distribution! Suppose the discrete distribution is too big to fit in memory, i.e. the typical streaming algorithm problem. The Gumbel trick can be used in this situation to draw samples with only **constant** memory requirements!

We also show an interesting equivalence of the Gumbel-max-trick to a well-known algorithm for weighted random sampling from a stream due to Efraimidis (2005).

Lastly, we show how Gumbel trick can be used from sampling without replacement.
